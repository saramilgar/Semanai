```python
##Ejercicio5
##Manipular un dataframe de strings
import numpy as np
arr= ["how", "to", "kick", "ass"]

def to_upper_case(s):
    return str(s).upper()
def print_iterator(it):
    for x in it:
        print(x,end= ' ')
        print(' ')
map_iterator = map(to_upper_case, arr)
print(type(map_iterator))
print_iterator(map_iterator)
```
